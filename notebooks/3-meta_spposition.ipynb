## 선수 포지션(spposition) 메타데이터 조회

> `@../docs/10-FCONLINE/14-Meta-Data.md`

### 개요

`GET /static/fconline/meta/spposition.json` 으로 전체 선수 포지션 목록을 수집한다.  
인증 불필요. 총 **29건** 수록.

| 컬럼 | 타입 | 원본 필드 | 변환 규칙 |
|---|---|---|---|
| `ID` | int | `spposition` | 그대로 |
| `role` | str | `desc` | 그대로 |

### ERD

```mermaid
erDiagram
    meta_spposition {
        INT ID PK
        STRING role
    }
```

### 저장 위치

```
data/meta_spposition.csv
```

````python
import csv
import os

import requests
from dotenv import load_dotenv

load_dotenv()

_BASE_URL = os.getenv("SERVERS")
_SPPOSITION_URL = f"{_BASE_URL}/static/fconline/meta/spposition.json"

OUTPUT = "../data/meta_spposition.csv"


def fetch_spposition_meta() -> list[dict]:
    resp = requests.get(_SPPOSITION_URL)
    resp.raise_for_status()
    return resp.json()


def parse_spposition(raw: list[dict]) -> list[dict]:
    return [
        {"ID": item["spposition"], "role": item["desc"]}
        for item in raw
    ]


def save_csv(records: list[dict], path: str) -> None:
    with open(path, "w", newline="", encoding="utf-8-sig") as f:
        writer = csv.DictWriter(f, fieldnames=["ID", "role"])
        writer.writeheader()
        writer.writerows(records)
    print(f"저장 완료: {path} ({len(records)}건)")


if __name__ == "__main__":
    print("spposition 메타데이터 수집 중...")
    raw = fetch_spposition_meta()
    records = parse_spposition(raw)
    print(f"총 {len(records)}건")
    save_csv(records, OUTPUT)
````
