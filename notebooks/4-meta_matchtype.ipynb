## 매치 종류(matchtype) 메타데이터 조회

> `@../docs/10-FCONLINE/14-Meta-Data.md`

### 개요

`GET /static/fconline/meta/matchtype.json` 으로 전체 매치 종류 목록을 수집한다.  
인증 불필요. 총 **9건** 수록.

| 컬럼 | 타입 | 원본 필드 | 변환 규칙 |
|---|---|---|---|
| `ID` | int | `matchtype` | 그대로 |
| `desc` | str | `desc` | 그대로 |

### ERD

```mermaid
erDiagram
    meta_matchtype {
        INT ID PK
        STRING desc
    }
```

### 저장 위치

```
data/meta_matchtype.csv
```

````python
import csv
import os

import requests
from dotenv import load_dotenv

load_dotenv()

_BASE_URL = os.getenv("SERVERS")
_MATCHTYPE_URL = f"{_BASE_URL}/static/fconline/meta/matchtype.json"

OUTPUT = "../data/meta_matchtype.csv"


def fetch_matchtype_meta() -> list[dict]:
    resp = requests.get(_MATCHTYPE_URL)
    resp.raise_for_status()
    return resp.json()


def parse_matchtype(raw: list[dict]) -> list[dict]:
    return [
        {"ID": item["matchtype"], "desc": item["desc"]}
        for item in raw
    ]


def save_csv(records: list[dict], path: str) -> None:
    with open(path, "w", newline="", encoding="utf-8-sig") as f:
        writer = csv.DictWriter(f, fieldnames=["ID", "desc"])
        writer.writeheader()
        writer.writerows(records)
    print(f"저장 완료: {path} ({len(records)}건)")


if __name__ == "__main__":
    print("matchtype 메타데이터 수집 중...")
    raw = fetch_matchtype_meta()
    records = parse_matchtype(raw)
    print(f"총 {len(records)}건")
    save_csv(records, OUTPUT)
````
